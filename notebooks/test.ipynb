```python
from pathlib import Path
import json

project_root = Path.cwd().parent

for ticker in ["AAPL", "MSFT", "AMZN"]:
    processed_dir = project_root / "data" / "processed" / ticker

    print("Checking:", processed_dir)

    files = list(processed_dir.glob("*_sections.json"))
    print(files)

    path = files[0]

    data = json.loads(path.read_text(encoding="utf-8"))

    print(f"\n===== {ticker} =====")

    for name, content in data["sections"].items():
        print(f"--- {name} ({len(content)} ký tự) ---")
        print(content[:250])
        print()
```
