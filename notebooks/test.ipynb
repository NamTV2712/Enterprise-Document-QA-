```python
from pathlib import Path
import re
from src.ingestion.section_extractor import SECTION_PATTERNS, html_to_text

project_root = Path.cwd().parent

html_path = project_root / "data" / "raw" / "MSFT" / "000095017025100235.html"
html = html_path.read_bytes()

text_lower = html_to_text(html).lower()

for name, pattern in SECTION_PATTERNS.items():
    positions = [m.start() for m in re.finditer(pattern, text_lower)]
    print(f"{name}: {len(positions)} lần xuất hiện -> {positions}")
```

```python
patterns = [
    "part i\nitem 1",
    "part i\nitem 1.",
    "item 1. business",
    "item 1.\nbusiness",
    "business\noverview",
    "our company",
]

for q in patterns:
    positions = [m.start() for m in re.finditer(re.escape(q), text_lower)]
    print(q, "->", positions[:20], "count =", len(positions))
```
